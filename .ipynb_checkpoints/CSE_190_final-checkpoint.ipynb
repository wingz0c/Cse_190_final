```python
import mido
from mido import MidiFile

# LUT for array indices storing note bits
channels = [0,1,2,9]
chanToArr = [0,1,2,0,0,0,0,0,0,3]

# octaves can be added if needed
note = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']


i = 1               # counter for each print statement
prevChan = 0        # keeps track of previous channel to detect channel changes
chanIt = 0          # iterates to the next channel in the forced order 0->1->2->9
arr = [[0]*12]*4    # 12 notes each for channels(0,1,2,9) to keep track of continued notes

# insert MIDI file here
mid = MidiFile('046_CastlevaniaIII_Dracula_sCurse_02_03Beginning.mid') 
temp = mido.merge_tracks(mid.tracks)

# increments modulo 4 to keep track of channel changes
def increment( a ):
    global i
    a = a + 1
    if(a == 4):
        a = a - 4
        print(str(i) + " NXT_STEP")
        i = i+1
    return a

# processing MIDI file
for msg in temp:
    # look for note changes
    if(msg.type=='note_on'):
        
        # check for a new channel
        if(msg.channel != prevChan): 
            chanIt = increment(chanIt)
            
            # iterate chanIt to "catch up" to the next listed channel
            while( channels[chanIt] != msg.channel):
                print(str(i) + " NXT_CHNL")
                
                # continue notes if not turned off
                for a in range(len(arr[chanToArr[msg.channel]])):
                    if(arr[chanToArr[msg.channel]][a]==1):
                        print(str(i) + " CNT_NOTE(" + str(a) + ")")
                        i = i+1
                
                chanIt = increment(chanIt)
                i = i+1
            
            print(str(i) + " NXT_CHNL")
            i = i+1

        
        # check if note_on event is on or off switch
        if(msg.velocity==0):
            noteSwitch = "off"
            arr[chanToArr[msg.channel]][int(msg.note)%12] = 0
        else:
            noteSwitch = "on"
            arr[chanToArr[msg.channel]][int(msg.note)%12] = 1        
            # print new note
            print(str(i) + " NEW_NOTE(" + note[int(msg.note) % 12] + ")") 
            i = i+1  

        # update prevChan for detection
        prevChan = msg.channel
```
